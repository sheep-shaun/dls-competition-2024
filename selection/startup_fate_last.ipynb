```python
import os

import pandas as pd
import numpy as np
```

```python
data_path = "./data/startup_fate/submissions"
files = os.listdir(data_path)
files = [file for file in files if file != "sub_6.csv"]
files
```

```python
data = pd.read_csv(os.path.join(data_path, "sub_0.csv"))
data["has_next_round"] = 0
data
```

```python
for file in files:
    sub = pd.read_csv(os.path.join(data_path, file))
    data["has_next_round"] += sub["has_next_round"]

data["has_next_round"] /= len(files)
data
```

```python
data["has_next_round"] = (data["has_next_round"] >= 0.5).astype(int)
data
```

```python
data.to_csv("./data/startup_fate/submissions/sub_7.csv", index=False)
```
